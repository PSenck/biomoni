```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
path = "../Examples/example_data/OPCUA/Results_2022-02-14_19-25-05/data_save.csv"
df = pd.read_csv(path, sep=";",encoding= "unicode_escape",decimal=",", skiprows=[1,2] , skipfooter=1, usecols = None, engine="python")
```

```python
df
noise1 = np.random.normal(0, 0.05, len(df))
noise2 = np.random.normal(0, 0.05, len(df))
noise1[noise1 < 0] = 0
for i in range(len(noise1)):
    if i > 0:
        if noise1[i] <= noise1[i-1]:
            noise1[i] = noise1[i-1]
#noise1 = noise1 - np.min(noise1)
df["BASET"] = df["BASET"] * (1+ noise1)
df["CO2"] = df["CO2"] * (1+ noise2)
```
