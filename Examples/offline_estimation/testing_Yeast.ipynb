```python
#Creating dict /list with several experiments and process some of them

experiment_dict = {exp : Experiment(path, exp) for exp in ["F4", "F5", "F6", "F7", "F8" ]}  #all experiments in a dictionary
experiment_dict["F8"].time_filter(dskey= "on", start = pd.to_datetime("14.12.2020  12:20:16")) #special time filter for experiment 8
[experiment_dict[exp].pop_dataframe("on") for exp in ["F4", "F5", "F6"]]   #delete online data in experiment 4,5,6 because of bad BASET_rate measurements
experiment_list = list(experiment_dict.values())   
```

```python
#instantiating yeast object from Yeast class
yeast = Yeast()
```

```python
#Change fit parameters to arbitrary to emonstrate estimation. Value have to be within min, max
yeast.change_params("qsmax", value =1, max = 10)
yeast.change_params("base_coef", value = 1)
yeast.change_params("qO2max", value = 0.2)
yeast.change_params("Yxs_ox" , value = 0.4)
display(yeast.p)
```

```python
#estimate parameters for given experiments in the list: You can give a list or dict of experiments or a single experiment
yeast.estimate(experiment_dict, tau = 1)     #it is possible to hand over a dict, a list (experiment_list for example) or a single experiment, it is also possible to hand over kwargs for solve_ivp in form of a dict and kwargs fpr lmfit.minimize , kwargs_solve_ivp = dict(method= "RK23", first_step = 0.0000001, max_step= 0.1), gtol = 0.1
##possible example  kwags arguments: max_nfev = 100, gtol = 0.1, xtol= 0.1 
```

```python
#show fit statistics
yeast.report()
```

```python
#Get datasets of experiments and simulate for each expriment, save both in seperate dictionaries
#it is necessary to estimate beforehand (yeast.estimate) for good simulation results, because yeast.p takes on the estimated params after estimation
t_grid = np.linspace(0,9,1001)      #artificial t_grid

sim_dict_all = {experiment.exp_id: yeast.simulate(experiment = experiment, t_grid = t_grid) for experiment in experiment_dict.values()}
#experiment_dataset_dict_all = {experiment.exp_id: experiment.dataset for experiment in experiment_dict.values()} # also possible to give

#visualize experiments and simulated data simultaneously
for exp_id  in experiment_dict.keys():
    title = "Experiment {0}".format(exp_id) 
    visualize(experiment_dict[exp_id] , sim_dict_all[exp_id], title = title, suffix_1= "_experimental", suffix_2 = "_fitted").show()
    
```

```python
exp_settings = {   "offline_est": dict(

    exp_id = "F7", meta_path = "metadata.xlsx"
    , types = {"off" : "offline.csv", "on": "online.CSV", "CO2" : "CO2.dat"}
    , index_ts = {"off" : 0, "on": 0, "CO2" : 0}

    , read_csv_settings = { "off" : dict(sep=";", encoding= 'unicode_escape', header = 0, usecols = None)
    , "on": dict(sep=";",encoding= "unicode_escape",decimal=",", skiprows=[1,2] , skipfooter=1, usecols = None, engine="python")
    , "CO2" : dict(sep=";", encoding= "unicode_escape", header = 0, skiprows=[0], usecols=[0,2,4], names =["ts","CO2", "p"])    }

    , to_datetime_settings = {"off" : dict(format = "%d.%m.%Y %H:%M", exact= False, errors = "coerce")
    , "on": dict(format = "%d.%m.%Y  %H:%M:%S", exact= False, errors = "coerce")
    , "CO2" : dict(format = "%d.%m.%Y %H:%M:%S", exact= False, errors = "coerce")   }

    , calc_rate = ("on", "BASET")
    , exp_dir_manual = None
    , endpoint = "end1"
    , read_excel_settings = None

    )
}


kwargs_estimate = {         #settings to estimate

    "online_est" : dict (
    tau = 1, 
    max_nfev = 100      #max function evaluations
    ),

    "offline_est" : dict ()
}

kwargs_exp = exp_settings["offline_est"]
kwargs_esti = kwargs_estimate["offline_est"]
```
